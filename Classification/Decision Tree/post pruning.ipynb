```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data
```

```python
iris.target
#setosa, versicolor, virginica
```

```python
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# independent and dependent features
x = df.iloc[:, :-1].values
y = iris.target
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
# training the model- postpruning. Hyperparameters are tuned after the model is built.
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (15, 10))
tree.plot_tree(treeclassifier, filled=True)
```

### Observation
In the root node, there are total 100 samples. 31 setosa, 35 versicolor, 34 virginica

if x[2] value <= 2.45, then the data point will mbe classified as setosa. that is why it says, 31, 0, 0. and gini value is 0 showing that it is a pure split.

In level 2, the left node says 0, 34, 4 which shows that the majority is versicolor. That is the ratio of versicolor is higher than then virginica. whereas the right node shows values as 0, 1, 30. Which again clearly shows that the majority is verginica. Theorefore it doesn't require any further split. We can simply set our max_depth=2. Because we need not split it any further. Otherwise our model will overfit.

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth = 2)
```

```python
treeclassifier.fit(x_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (15, 10))
tree.plot_tree(treeclassifier, filled=True)
```

Level is set to 2 and we don't have any further splits. Overfitting is reduced.

```python
y_pred = treeclassifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred, y_test))
```
